## Part 1: The Mental Model (Read this first\!)

*Before writing code, you must understand the hierarchy.*

In Matplotlib, we distinguish between two main objects:

1.  **The Figure (`fig`):** The blank canvas or the "page". It holds everything.
2.  **The Axes (`ax`):** The actual plot/graph. A Figure can have multiple Axes (subplots).

*Think of it like painting: You buy one Canvas (`fig`), and you paint two separate Landscapes (`ax1`, `ax2`) on it.*


## Part 2: Setup & The "Hello World"

*Let's import libraries and generate some dummy data.*

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure charts display inline in Jupyter
%matplotlib inline 

# Generate Dummy Data
# x: 100 points from 0 to 10
x = np.linspace(0, 10, 100) 
# y: Sine wave
y = np.sin(x)

# --- THE "PROFESSIONAL" SYNTAX ---
# 1. Create the Objects (The Container and the Plot)
fig, ax = plt.subplots() 

# 2. Add Data to the Axes
ax.plot(x, y)

# 3. Show the result
plt.show()
```


## Part 3: Styling & Labeling

*A chart without labels is useless. Here is how to annotate.*

```python
# Re-generate data for independence
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# 1. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 5)) # figsize=(Width, Height) in inches

# 2. Plot Multiple Lines
# We can chain commands or call ax.plot multiple times
ax.plot(x, y, color='blue', linestyle='--', label='Sine Wave')
ax.plot(x, z, color='red',  linestyle='-',  label='Cosine Wave')

# 3. Set Labels (Note the 'set_' prefix common in OO-style)
ax.set_title("Trigonometric Functions")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude")

# 4. Add Legend (Uses the 'label' arguments defined above)
ax.legend()

# 5. Add Grid
ax.grid(True, alpha=0.3) # alpha = transparency

plt.show()
```

## Part 4: Common Plot Types

*Line charts are just the beginning.*

### 4.1 Scatter Plots (Relationships)

*Used to see correlation between two variables.*

```python
# Random Data
np.random.seed(42)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

fig, ax = plt.subplots()

# Scatter: x, y, s=Size, c=Color, alpha=Transparency
ax.scatter(x_scatter, y_scatter, s=sizes, c=colors, alpha=0.5, cmap='viridis')

ax.set_title("Bubble Chart / Scatter Plot")
plt.show()
```

### 4.2 Bar Charts (Categorical)

*Used for comparing categories.*

```python
categories = ['Apple', 'Banana', 'Cherry']
values = [10, 24, 36]

fig, ax = plt.subplots()

# Bar Chart
ax.bar(categories, values, color=['green', 'yellow', 'red'])

ax.set_title("Fruit Prices")
plt.show()
```

### 4.3 Histograms (Distributions)

*Used to see the frequency distribution of data.*


```python
# Generate 1000 data points from a Normal Distribution
data = np.random.randn(1000)

fig, ax = plt.subplots()

# Hist: bins=Number of bars
ax.hist(data, bins=30, color='purple', edgecolor='black', alpha=0.7)

ax.set_title("Normal Distribution Histogram")
plt.show()
```

## Part 5: Subplots (Multiple Charts)

*This is where the Object-Oriented approach shines. We create a grid of Axes.*


```python
x = np.linspace(0, 10, 100)

# 1. Create a 2x2 Grid (2 Rows, 2 Columns)
# This returns 'ax' as a 2D array of Axes objects
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# 2. Access each subplot by index [row, col]
# Top Left
axes[0, 0].plot(x, np.sin(x), 'g')
axes[0, 0].set_title("Sine")

# Top Right
axes[0, 1].plot(x, np.cos(x), 'b')
axes[0, 1].set_title("Cosine")

# Bottom Left
axes[1, 0].plot(x, np.tan(x), 'r')
axes[1, 0].set_title("Tangent")

# Bottom Right (Let's leave it blank or plot something else)
axes[1, 1].hist(np.random.randn(100), bins=20)
axes[1, 1].set_title("Noise")

# 3. Clean up layout (Prevents labels overlapping)
plt.tight_layout()

plt.show()
```


## Part 6: Saving Figures

*How to export your graph for a report or presentation.*

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_title("Saved Image")

# Save to the current directory
# dpi = Dots Per Inch (Resolution). 300 is standard for print.
fig.savefig("my_chart.png", dpi=300, bbox_inches='tight')

print("Image saved as my_chart.png")
```

## Part 7: Final Exercise - The "Stock Price" Dashboard

*Combine everything: Time series, Fill between, and Annotations.*

```python
# Generate Dummy Stock Data
days = np.arange(0, 100)
price = np.cumsum(np.random.randn(100)) + 100 # Random walk starting at 100

fig, ax = plt.subplots(figsize=(12, 6))

# 1. Main Line
ax.plot(days, price, color='black', linewidth=2, label='Stock Price')

# 2. Fill Below (Area chart effect)
ax.fill_between(days, price, alpha=0.2, color='skyblue')

# 3. Highlight the Peak
max_price = np.max(price)
max_day = days[np.argmax(price)]

# Annotate: Adds text with an arrow
ax.annotate(f'Peak: ${max_price:.2f}', 
            xy=(max_day, max_price),           # Point to annotate
            xytext=(max_day+10, max_price+5),  # Text position
            arrowprops=dict(facecolor='red', shrink=0.05))

# 4. Styling
ax.set_title("Simulated Stock Performance")
ax.set_xlabel("Days")
ax.set_ylabel("Price ($)")
ax.grid(True, linestyle='--')
ax.legend()

plt.show()
```
